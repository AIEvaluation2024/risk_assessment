```python
import numpy as np
from matplotlib import pyplot as plt

data_types = ["California_Housing","CCPP","Kin8","Naval_Propulsion","WineWhite"]
legends = ["CH","CCPP","K8","NP","WW"]

train_loss = []
for count,data in enumerate(data_types):
    train_loss = np.genfromtxt("training_loss_" + "NN" + "_split_" + data + ".csv")
    plt.plot(range(len(train_loss)),np.log(train_loss),label = legends[count])
plt.grid(True)
plt.legend()
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('epochs',fontsize = 14)
plt.ylabel('log(MSE)',fontsize = 14)
plt.title('Loss on test set for NN-1')
plt.show()
```
